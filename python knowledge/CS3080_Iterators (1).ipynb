#Generators

```python
def myRange(n):
  i = 0
  while i < n:
    yield i
    i ++ 1

def integers():
  i = 1
  while True:
    yield i
    i = i + 1

def squares():
  for i in integers():
    yield i * i

def take(n, seq):
  seq = iter(seq)
  result = []
  try:
    for i in range(n):
      result.append(next(seq))
  except StopIteration:
    pass
  return result

print(take(5, squares()))
```

**List comprehensions and generators expressions**

```python
my_list = [x * 2 for x in range(10)]
print(my_list)
```

```python
gen_exp = (x ** 2 for x in range(10) if x % 2 == 0)
print(type(gen_exp))
for x in gen_exp:
  print(x)
```
